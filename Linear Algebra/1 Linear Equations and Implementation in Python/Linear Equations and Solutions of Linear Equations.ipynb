## Introduction to Linear Equations

Linear equations are fundamental mathematical expressions that describe the relationship between variables in a straight-line form. The general form of a linear equation in one variable is:

\[ ax + b = 0 \]

where \(a\) and \(b\) are constants, and \(x\) is the variable.

In the case of linear equations in two variables, the general form is:

\[ ax + by + c = 0 \]

Here, \(a\), \(b\), and \(c\) are constants, and \(x\) and \(y\) are the variables.

## Solving Linear Equations

### Example: Solving a System of Linear Equations in Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Coefficients matrix
coefficients = np.array([[2, 3], [4, -2]])

# Constants vector
constants = np.array([8, 6])

# Solve for variables
solution = np.linalg.solve(coefficients, constants)

print("Solution:")
print("x =", solution[0])
print("y =", solution[1])
```

```python
# Plotting
x_values = np.linspace(-5, 5, 100)

# Equations for plotting
equation1 = (constants[0] - coefficients[0, 0] * x_values) / coefficients[0, 1]
equation2 = (constants[1] - coefficients[1, 0] * x_values) / coefficients[1, 1]

# Plot the equations
plt.figure(figsize=(8, 6))
plt.plot(x_values, equation1, label='2x + 3y = 8')
plt.plot(x_values, equation2, label='4x - 2y = 6')
plt.scatter(solution[0], solution[1], color='red', label='Intersection Point')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('System of Linear Equations')
plt.legend()
plt.grid(True)
plt.show()
```

a_{21}x_1 + a_{22}x_2 + \ldots + a_{2n}x_n &= b_2 \\
&\vdots \\
a_{m1}x_1 + a_{m2}x_2 + \ldots + a_{mn}x_n &= b_m \\
\end{align*} \]

Here, \(a_{ij}\) are coefficients, \(x_i\) are variables, and \(b_i\) are constants for each equation. Solving a system of linear equations means finding values for the variables that satisfy all equations simus)

## Methods of Solution

### 1. **Substitution Method:**
   - Solve one equation for one variable and substitute it into the other equations.
   - Repeat until all variables are determined.

   **Example:**
   \[ \begin{align*}
   2x + y &= 5 \\
   3x - 2y &= 8 \\
   \end{align*} \]

   Solve the first equation for \(y\): \(y = 5 - 2x\)

   Substitute into the second equation: \(3x - 2(5 - 2x) = 8\)



```python
# Substitution Method
from sympy import symbols, Eq, solve

# Define variables
x, y = symbols('x y')

# Given equations
eq1 = Eq(2*x + y, 5)
eq2 = Eq(3*x - 2*y, 8)

# Solve for y in terms of x
y_solution = solve(eq1, y)[0]

# Substitute y_solution into eq2
solution = solve(eq2.subs(y, y_solution), x)

print("Substitution Method:")
print("Solution for x:", solution[0])
print("Solution for y:", y_solution.subs(x, solution[0]))
```

### 2. **Elimination Method:**
   - Multiply equations to eliminate one variable when added or subtracted.
   - Solve for the remaining variable.

   **Example:**
   \[ \begin{align*}
   2x + 3y &= 8 \\
   4x - 2y &= 6 \\
   \end{align*} \]

   Multiply the first equation by 2 to eliminate \(y\): \(4x + 6y = 16\)

   Subtract the second equation: \((4x + 6y) - (4x - 2y) = 16 - 6\)



```python
from sympy import symbols, Eq, solve

# Define symbols
x, y = symbols('x y')

# Elimination Method
# Given equations
eq3 = Eq(2*x + 3*y, 8)
eq4 = Eq(4*x - 2*y, 6)

# Multiply eq3 by 2
eq3_multiplied = Eq(eq3.lhs * 2, eq3.rhs * 2)

# Subtract eq4 from eq3_multiplied
eq5 = Eq(eq3_multiplied.lhs - eq4.lhs, eq3_multiplied.rhs - eq4.rhs)

try:
    # Solve for x
    x_solution = solve(eq5, x)[0]

    # Substitute x_solution into eq3
    y_solution = solve(eq3.subs(x, x_solution), y)

    print("\nElimination Method:")
    print("Solution for x:", x_solution)
    print("Solution for y:", y_solution[0])
except IndexError:
    print("\nNo solution found for x. The system may be inconsistent or the solution space is empty.")
```

# Gaussian Elimination Method

Gaussian Elimination is a systematic method used for solving a system of linear equations by transforming the augmented matrix into its row-echelon or reduced row-echelon form. The objective is to simplify the system of equations and make it easier to find the solution.

## Steps of Gaussian Elimination:

### Step 1: Augmented Matrix
Start with the augmented matrix, which includes both the coefficients matrix and the constants vector.

### Step 2: Row Operations
Perform row operations to transform the matrix into upper triangular form. The three elementary row operations are:
- Multiply a row by a nonzero scalar.
- Add or subtract one row from another.
- Swap the positions of two rows.

### Step 3: Back Substitution
Once the matrix is in upper triangular form, perform back substitution to find the values of variables.

## Example:

Consider the system of linear equations:
\[ 
\begin{align*}
2x + 3y &= 8 \\
4x - 2y &= 6
\end{align*}
\]

### Step 1: Augmented Matrix
\[ 
\begin{bmatrix}
2 & 3 & | & 8 \\
4 & -2 & | & 6
\end{bmatrix}
\]

### Step 2: Row Operations
Subtract 2 times the first row from the second row to eliminate the coefficient of \(x\) in the second equation.

\[ 
\begin{bmatrix}
2 & 3 & | & 8 \\
0 & -8 & | & -10
\end{bmatrix}
\]

### Step 3: Back Substitution
Solve for \(y\) in the second equation: \(-8y = -10 \implies y = \frac{5}{4}\).

Substitute \(y\) back into the first equation: \(2x + 3\left(\frac{5}{4}\right) = 8\).

Solve for \(x\): \(2x + \frac{15}{4} = 8 \implies x = \frac{17}{4}\).

Therefore, the solution is \(x = \frac{17}{4}\) and \(y = \frac{5}{4}\).

Gaussian Elimination is a powerful method for solving systems of linear equations and is widely used in numerical linear algebra and computational mathematics.


```python
from sympy import symbols, Eq, solve, Matrix

# Define symbols
x, y, z = symbols('x y z')

# Example 1:
# System of linear equations: 2x + 3y = 8, 4x - 2y = 6
eq1 = Eq(2*x + 3*y, 8)
eq2 = Eq(4*x - 2*y, 6)

# Augmented matrix
augmented_matrix1 = Matrix([[2, 3, 8], [4, -2, 6]])

# Apply Gaussian Elimination
reduced_matrix1, pivot_columns1 = augmented_matrix1.rref()
print(reduced_matrix1)
print(pivot_columns1)

# Solve for variables
solution1 = solve((eq1, eq2), (x, y))
print("Example 1 Solution:", solution1)
```

```python
import numpy as np

def gaussian_elimination(A, B):
    # Explicitly cast to float64
    A = A.astype(np.float64)
    B = B.astype(np.float64)

    # Combine the coefficient matrix A and the constants matrix B
    augmented_matrix = np.concatenate((A, B.reshape(-1, 1)), axis=1)
    rows, cols = augmented_matrix.shape

    for pivot_row in range(rows):
        # Make the diagonal element (pivot) equal to 1
        augmented_matrix[pivot_row] /= augmented_matrix[pivot_row, pivot_row]

        # Eliminate non-zero values below and above the pivot
        for i in range(rows):
            if i != pivot_row:
                factor = augmented_matrix[i, pivot_row]
                augmented_matrix[i] -= factor * augmented_matrix[pivot_row]

    # Extract the solutions from the rightmost column
    solutions = augmented_matrix[:, -1]

    return solutions
```

```python
# Example 1:
A1 = np.array([[2, 3], [4, -2]])
B1 = np.array([8, 6])
solution1 = gaussian_elimination(A1, B1)
print("Example 1 Solution:", solution1)
```

```python
# Example 2:
A2 = np.array([[1, 1, 1], [0, 2, -1], [2, 5, -1]])
B2 = np.array([6, 3, 9])
solution2 = gaussian_elimination(A2, B2)
print("Example 2 Solution:", solution2)
```

```python
# Example 3:
A3 = np.array([[1, 1, -1], [2, -1, 1], [3, 2, 1]])
B3 = np.array([0, 3, 9])
solution3 = gaussian_elimination(A3, B3)
print("Example 3 Solution:", solution3)
```

# Consistent or Inconsistent System of Linear Equations

A system of linear equations is considered **consistent** if it has at least one solution, and it is deemed **inconsistent** if it has no solution. Let's explore the conditions for consistency and inconsistency.

## Consistent System

### Unique Solution
A system is said to have a **unique solution** if there exists only one set of values for the variables that satisfies all the equations. Geometrically, this corresponds to the intersection of multiple planes (in three dimensions) or lines (in two dimensions) at a single point.

### Infinitely Many Solutions
A system is also considered consistent if it has **infinitely many solutions**. This occurs when the equations represent parallel planes or lines that overlap. The overlapping region represents an infinite set of solutions.

## Inconsistent System

An **inconsistent system** has no solution, meaning there are contradictory constraints among the equations. In three dimensions, this corresponds to three planes intersecting at a point that does not exist. In two dimensions, it could be parallel lines that never intersect.

## Conditions for Consistency

### Number of Equations and Variables
A system with fewer equations than variables is typically underdetermined and has infinitely many solutions. A system with more equations than variables may be overdetermined, leading to either a unique solution, no solution, or infinitely many solutions.

### Rank of Coefficient Matrix
The **rank of the coefficient matrix** plays a crucial role. If the rank of the coefficient matrix is equal to the rank of the augmented matrix, the system is consistent. If the ranks are unequal, the system is inconsistent.

### Row Echelon Form
Transforming the system to **row-echelon form** or **reduced row-echelon form** can help identify the consistency. If the last column of the row-echelon form contains a pivot position, the system is inconsistent.

## Example:

Consider the system of equations:

\[ \begin{align*}
2x + 3y &= 8 \\
4x - 2y &= 6 \\
\end{align*} \]

This system is consistent and has a unique solution. The coefficient matrix's rank is 2, and the augmented matrix's rank is also 2.

Now, consider a modified system:

\[ \begin{align*}
2x + 3y &= 8 \\
4x - 6y &= 12 \\
\end{align*} \]

This system is inconsistent because the coefficients in the second equation result in a contradiction. The coefficient matrix's rank is 2, but the augmented matrix's rank is 1.

Understanding the consistency of a system is crucial for solving real-world problems and ensuring the validity of the solutions.


```python
import numpy as np

def is_consistent(A, B):
    # Combine the coefficient matrix A and the constants matrix B
    augmented_matrix = np.concatenate((A, B.reshape(-1, 1)), axis=1)

    # Check the rank of the coefficient matrix and the augmented matrix
    rank_A = np.linalg.matrix_rank(A)
    rank_augmented = np.linalg.matrix_rank(augmented_matrix)

    # Check if the system is consistent
    return rank_A == rank_augmented
    
```

```python
# Example 1: Consistent System with Unique Solution
A1 = np.array([[2, 3], [4, -2]])
B1 = np.array([8, 6])
consistent1 = is_consistent(A1, B1)
print("Example 1 is consistent:", consistent1)

```

```python
# Example 2: Inconsistent System
A2 = np.array([[2, 3], [4, -6]])
B2 = np.array([8, 12])
consistent2 = is_consistent(A2, B2)
print("Example 2 is consistent:", consistent2)
```

```python
from sympy import Matrix

def is_consistent_sympy(A, B):
    # Combine the coefficient matrix A and the constants matrix B
    augmented_matrix = Matrix.hstack(Matrix(A), Matrix(B))

    # Check the rank of the coefficient matrix and the augmented matrix
    rank_A = augmented_matrix[:, :-
    1].rank()
    rank_augmented = augmented_matrix.rank()

    # Check if the system is consistent
    if rank_A == rank_augmented:
        return True
    else:
        return False

```

```python
# Example 1: Consistent System with Unique Solution
A1 = [[2, 3], [4, -2]]
B1 = [8, 6]
consistent1 = is_consistent_sympy(A1, B1)
print("Example 1 is consistent:", consistent1)
```

```python
# Example 2: Inconsistent System
A2 = [[2, 3], [4, -6]]
B2 = [8, 12]
consistent2 = is_consistent_sympy(A2, B2)
print("Example 2 is consistent:", consistent2)
```

```python
from sympy import Matrix

def check_consistency(A, B):
    # Combine the coefficient matrix A and the constants matrix B
    augmented_matrix = Matrix.hstack(Matrix(A), Matrix(B))

    # Find the rank of the coefficient matrix and the augmented matrix
    rank_A = augmented_matrix[:, :-1].rank()
    rank_augmented = augmented_matrix.rank()

    # Check if the system is consistent
    if rank_A == rank_augmented:
        if rank_A == len(A[0]):
            return "The system is consistent and has a unique solution."
        else:
            return "The system is consistent and has infinitely many solutions."
    else:
        return "The system is inconsistent and has no solution."

```

```python
# Example 1: Consistent System with Unique Solution
A1 = [[2, 3], [4, -2]]
B1 = [8, 6]
result1 = check_consistency(A1, B1)
print("Example 1:", result1)

```

```python
# Example 2: Inconsistent System
A2 = [[2, 3], [4, -6]]
B2 = [8, 12]
result2 = check_consistency(A2, B2)
print("Example 2:", result2)
```
